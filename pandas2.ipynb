```python
graph_df= pd.DataFrame()

for region in df['region'].unique():
    print(region)
    region_df= df.copy()[df['region'] == region]
    
    region_df.set_index("Date", inplace=True)
    region_df.sort_index(inplace=True)
    region_df[f'{region}_price_25ma']=region_df['AveragePrice'].rolling(25).mean()
    
    
   
   
    if graph_df.empty:
        graph_df= region_df[[f'{region}_price_25ma']]
    else:
        graph_df= graph_df.join(region_df[f'{region}_price_25ma'])
        
graph_df.tail()
        
```
